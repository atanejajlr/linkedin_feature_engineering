# Removal of outliers using IQR (see references section for the link to the dataset)

You are given the height weight dataset which cmprises of heights and weights of 1000 people. Dataset is taken from here: https://www.kaggle.com/datasets/mustafaali96/weight-height

You need to do the following:

(1) Load the csv in the pandas dataframe and first plot the histograms for height and weight parameters

(2) Using IQR, detect weight outliers and pprint them

(3) Using IQR, detect height outliers and print them

```python
#Load the dataset

import pandas as pd

df = pd.read_csv("height_weight.csv")
```

```python
df.sample(20)
```

```python
#Now, we have to plot the histogram

import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10,6)
```

```python
# To plot the histogram of heights
# Histogram is the graphical display of features and is displayed using bars
# Here, we have to plot the histogram for heights and weights
# Below, we plot the histogram for heights

plt.hist(df.height, bins=25, rwidth=0.8)
plt.xlabel("height in inches")
plt.ylabel("Count")
plt.show()
```

```python
# To plot the histogram of weights
# Histogram is graphical display of features and is displayed using bars
# Here we plot the histogram of heights and weights
# Below we plot the histogram for weights
plt.hist(df.weight, bins=25, rwidth=0.8)
plt.xlabel('weight in pounds')
plt.ylabel('Count')
plt.show()
```

#Next using IQR detect weight outliers and print them

Recalling the IQR definition - this plot tells us all !



```python
Q1 = df.weight.quantile(0.25)
Q3 = df.weight.quantile(0.75)
IQR = Q3 - Q1

min_threshold = Q1 - 1.5*IQR
max_threshold = Q3 + 1.5*IQR
df1 = df[(df.weight < min_threshold) | (df.weight > max_threshold)]
df1.weight
```

```python
Q1 = df.height.quantile(0.25)
Q3 = df.height.quantile(0.75)
IQR = Q3 - Q1

min_threshold = Q1 - 1.5*IQR
max_threshold = Q3 + 1.5*IQR
df1 = df[(df.height < min_threshold) | (df.height > max_threshold)]
df1.height
```
