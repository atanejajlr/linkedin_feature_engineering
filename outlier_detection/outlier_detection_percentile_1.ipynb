# About this notebook

1) This jupyter notebook shows the heights database. We can see by observation that there are definitely some outliers in the dataset comprising of heights of erroneous values which cannot eb correct. 

2) We divide the datasets into percentiles/quantiles Q1, Q2, Q3,Q4 or 25, 50, 875 and 100 percentiles 

3) The minimum values is the 0th percentile and the maximum value is the 100th percentile

4) The IQR is calculated as Q3-Q1

5) The boxplot below helps understand about Q1/Q2/Q3/Q4



6) The lower and the upper limit to eliminate the outliers (beyond ths limit) is calculated as Q1 - 1.5 IQR and Q3 + 1.15 IQR

7) The steps are shown in the code below.



```python
#Load the data
import pandas as pd
df = pd.read_csv("heights.csv")
df.head()

```

```python
#Let's see the height column

df['height']
```

8) Now, lets lets see the height value which is 95th percentile

9) You will notice that the height of 9.68 is the 95th pecentile.

10) Anything above this may be considered as an outlier

11) We will set the maximum threshold as the 95th percentile.

12) This value of threshold should come from domain knowledge- there is no rule regarding the maximum threshold value

13) We will set the minimum threshold as 5th percentile

14) We will then get a new dataframe with these outliers ermoved

15) Again, there i no rule for the max and minimum threshold for removing the outliers and it has to come from the domain knowledge


```python
df['height'].quantile(0.95)
```

```python
max_threshold = df['height'].quantile(0.95)
max_threshold
```

```python
#Let's get height values greater than the max_threshold

df[df['height'] > max_threshold]
```

```python
min_threshold = df['height'].quantile(0.05)
min_threshold
```

```python
#Now, let's get the height value less than the 5th percentile

df[df['height'] < min_threshold]
```

```python
# Now lets geta new dataframe without the outliers

df[(df['height']<max_threshold) & (df['height'] > min_threshold)]
```

```python
#Thus, above we have got the dataframe with heights within the limits of max and min threshold.
```
